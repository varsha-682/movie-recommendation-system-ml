```python
# Import required libraries
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# Step 1: Create movie dataset
data = {
    "Movie": [
        "Inception","Interstellar","The Dark Knight","Titanic","Avatar",
        "The Avengers","Iron Man","Doctor Strange","Spider Man",
        "Joker","Batman Begins","Avengers Endgame","Frozen",
        "The Lion King","Aladdin","Toy Story","Finding Nemo"
    ],

    "Genre": [
        "Sci-Fi Thriller","Sci-Fi Drama","Action Crime","Romance Drama","Sci-Fi Adventure",
        "Action Sci-Fi","Action Sci-Fi","Fantasy Action","Action Adventure",
        "Crime Drama","Action Crime","Action Sci-Fi","Animation Musical",
        "Animation Adventure","Fantasy Adventure","Animation Comedy","Animation Adventure"
    ]
}

# Step 2: Convert dataset into DataFrame
movies = pd.DataFrame(data)

print("Movie Dataset:\n")
print(movies)


# Step 3: Convert genres into numerical vectors
cv = CountVectorizer()
matrix = cv.fit_transform(movies["Genre"])


# Step 4: Calculate similarity between movies
similarity = cosine_similarity(matrix)


# Step 5: Recommendation function
def recommend(movie):

    movie_index = movies[movies["Movie"] == movie].index[0]

    distances = similarity[movie_index]

    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    print("\nRecommended Movies:")

    for i in movie_list:
        print(movies.iloc[i[0]].Movie)


# Step 6: Take user input
movie_name = input("\nEnter a movie name: ")

recommend(movie_name)
```
